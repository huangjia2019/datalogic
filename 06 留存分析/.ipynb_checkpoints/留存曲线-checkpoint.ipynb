```python
import numpy as np # NumPy
import pandas as pd # Pandas
df_member = pd.read_csv('.csv') # 
df_member.head() # 
```

```python
df_member[''].replace(to_replace='', value=1, inplace=True) # 1
df_member[''].replace(to_replace='', value=0, inplace=True) # 0
```

```python
df_member[''].replace(to_replace='', value=0, inplace=True) # -0
df_member[''].replace(to_replace='', value=1, inplace=True) # -1
# 
binary_features = ['', '', '', '', '', '','A','B']
for field in binary_features:
    df_member[field] = df_member[field] == ''
# 
df_member[''] = pd.to_numeric(df_member[''], errors='coerce')
df_member[''].fillna(0, inplace=True)
```

```python
print(df_member.info())
```

```python
# pip install lifelines
```

```python
import lifelines # Lifelines
import matplotlib.pyplot as plt # pyplot 
plt.rcParams["font.family"]=['SimHei'] #
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.rcParams['axes.unicode_minus']=False #
kmf = lifelines.KaplanMeierFitter() # KaplanMeier 
kmf.fit(df_member[''], event_observed=df_member[''], label='') #
fig, ax = plt.subplots(figsize=(10,6)) # 
kmf.plot(ax=ax,ci_alpha=0) # 
ax.set_title('Kaplan-Meier - ') # 
ax.set_xlabel('') #x 
ax.set_ylabel('(%)') #y 
```

```python
def life_by_cat(feature, t='', event='', df=df_member, ax=None): # 
    for cat in df[feature].unique(): # 
        idx = df[feature] == cat # 
        kmf = lifelines.KaplanMeierFitter() # KaplanMeier 
        kmf.fit(df[idx][t], event_observed=df[idx][event], label=cat) # 
        kmf.plot(ax=ax, label=cat,ci_alpha=0) # 
```

```python
fig_pmt, ax_pmt = plt.subplots(figsize=(10,6)) # 
life_by_cat(feature='', ax=ax_pmt) # 
ax_pmt.set_title('') # 
ax_pmt.set_xlabel('') #x 
ax_pmt.set_ylabel('(%)') #y 
```

```python
fig_contract, ax_contract = plt.subplots(figsize=(10,6)) # 
life_by_cat(feature='', ax=ax_contract) # 
ax_contract.set_title('') # 
ax_contract.set_xlabel('') #x 
ax_contract.set_ylabel('(%)') #y 
```

```python
fig_gender, ax_gender = plt.subplots(figsize=(10,6)) # 
life_by_cat(feature='', ax=ax_gender) # 
ax_gender.set_title('') # 
ax_gender.set_xlabel('') #x 
ax_gender.set_ylabel('(%)') #y 
```

```python
#
category_features = ['', ''] #
df_member = pd.get_dummies(df_member, #1 
                           drop_first=True, #drop_first=True
                           columns=category_features) # 
df_member #
```

```python
cph = lifelines.CoxPHFitter() # CoxPH 
cph.fit(df_member, duration_col='', event_col='', show_progress=False) # 
cph.print_summary() # 
```

```python
fig_coef, ax_coef = plt.subplots(figsize=(12,7)) # 
ax_coef.set_title('') # 
cph.plot(ax=ax_coef); # 
```

```python
X = df_member.drop(['',''], axis = 1) # 
y = df_member. .values # 
```

```python
from sklearn.model_selection import train_test_split # train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2) # 
```

```python
from sklearn.linear_model import LogisticRegression # 
logreg = LogisticRegression() # 
```

```python
logreg.fit(X_train,y_train) # 
print("SK-learn {:.2f}%".format(logreg.score(X_test,y_test)*100)) # 
```

```python
y_pred = logreg.predict(X_test) # 
print(" ", y_pred[0]) # 
```
