# 数据分析咖哥十话

## 第9话 君向潇湘我向秦：A/B测试助力促销

**题解** 本话的标题是一句唐诗*，借用郑谷和友人前路的两个方向，暗喻A / B 测试的两个实验方案。当然，郑谷和友人一个向南，一个向北，方向不同；而我们这里要讲的A / B 测试通常用于比较两个有细微差异的方案的优劣。



AB测试示意
详细内容请参考拙作：《数据分析咖哥十话》 人民邮电出版社2022年出版

```python
import pandas as pd # 导入Pandas
import numpy as np # 导入NumPy
import matplotlib.pyplot as plt # 导入pyplot 模块

plt.rcParams["font.family"]=['SimHei'] #用来设定字体样式
plt.rcParams['font.sans-serif']=['SimHei'] #用来设定无衬线字体样式
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
df_AB = pd.read_csv('AB测试.csv') # 载入数据
print('用户数:', df_AB.count()['用户码']) # 数据条目
df_AB.head() # 输出前几行数据
```

```python
ax = df_AB.groupby('页面版本').count()['用户码'].plot.pie(autopct='%1.0f%%') # 饼图
plt.show() # 输出饼图
```

```python
ax = df_AB.groupby(['页面版本', '是否购买']).count()['用户码'].unstack('是否购买').plot(kind='bar') #分组的转化条形图
plt.show() # 输出条形图
```

```python
new_conversion = df_AB.loc[(df_AB['是否购买'] == '是') &
(df_AB['页面版本'] == '新页面')].count()['用户码']
old_conversion = df_AB.loc[(df_AB['是否购买'] == '是') &
(df_AB['页面版本'] == '旧页面')].count()['用户码']
user_count = df_AB.count()['用户码'] * 0.5
print('新页面转化率：', (new_conversion / user_count))
print('旧页面转化率：', (old_conversion / user_count))
```

```python
from scipy import stats #导入Scipy中的stats模块
```

```python
def eval_AB_test(test_group, control_group): #定义函数
    AB_test_result = stats.ttest_ind(test_group, control_group) #返回P值，t值
    print('P值：', AB_test_result[1]) #P值
    print('t值：', AB_test_result[0]) #t值
    if AB_test_result [1] < 0.05: #以5%为阈值返回结果是否可信
        print('结果可信')
    else:
        print('结果不可信')
```

```python
df_AB['是否购买'].replace(to_replace='是', value=1, inplace=True) # 用“1”替换“是”
df_AB['是否购买'].replace(to_replace='否', value=0, inplace=True) # 用“0”替换“否”
df_test = df_AB.loc[(df_AB['页面版本'] == '新页面'), '是否购买'] # 测试组数组
df_control = df_AB.loc[(df_AB['页面版本'] == '旧页面'), '是否购买'] # 控制组数组
```

```python
eval_AB_test(df_test, df_control) # 调用函数
```

```python
df_test = df_AB.loc[(df_AB['页面版本'] == '新页面')&(df_AB['价值组别'] == '低'), '是否购买'] # 测试组数组( 低价值客户)
df_control = df_AB.loc[(df_AB['页面版本'] == '旧页面')&(df_AB['价值组别'] == '低'), '是否购买']# 测试组数组( 高价值客户)
```

```python
eval_AB_test(df_test, df_control) # 调用函数
```
