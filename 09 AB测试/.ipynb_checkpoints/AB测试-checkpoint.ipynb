```python
new_conversion = df_AB.loc[(df_AB[''] == '') &
(df_AB[''] == '')].count()['']
old_conversion = df_AB.loc[(df_AB[''] == '') &
(df_AB[''] == '')].count()['']
user_count = df_AB.count()[''] * 0.5
print('', (new_conversion / user_count))
print('', (old_conversion / user_count))
```

```python
from scipy import stats #Scipystats
```

```python
def eval_AB_test(test_group, control_group): #
    AB_test_result = stats.ttest_ind(test_group, control_group) #Pt
    print('P', AB_test_result[1]) #P
    print('t', AB_test_result[0]) #t
    if AB_test_result [1] < 0.05: #5%
        print('')
    else:
        print('')
```

```python
df_AB[''].replace(to_replace='', value=1, inplace=True) # 1
df_AB[''].replace(to_replace='', value=0, inplace=True) # 0
df_test = df_AB.loc[(df_AB[''] == ''), ''] # 
df_control = df_AB.loc[(df_AB[''] == ''), ''] # 
```

```python
eval_AB_test(df_test, df_control) # 
```

```python
df_test = df_AB.loc[(df_AB[''] == '')&(df_AB[''] == ''), ''] # ( )
df_control = df_AB.loc[(df_AB[''] == '')&(df_AB[''] == ''), '']# ( )
```

```python
eval_AB_test(df_test, df_control) # 
```
